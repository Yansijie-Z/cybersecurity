```python
data = pd.read_csv('KDDTrain+.txt', header=None)
ConnectionStateType = data.iloc[:, 3]
classification_counts = ConnectionStateType.value_counts()
total_count = len(ConnectionStateType)
classification_probabilities = classification_counts / total_count

# Step 4: Combine counts and probabilities into a DataFrame for better visualization
classification_summary = pd.DataFrame({
    'Count': classification_counts,
    'Probability': classification_probabilities
})

classification_summary.to_csv('ConnectionStateType_analysis.csv', index=True)
```

```python
data = {
    'Model': ['RF', 'SVM linear', 'LR', 'KNN', 'AdaBoost', 'DT', 'NB'],
    'Accuracy': [0.998, 0.998, 0.992, 0.998, 0.848, 0.998, 0.528],
    'Precision': [0.998, 0.998, 0.992, 0.998, 0.848, 0.998, 0.528],
    'Recall': [0.998, 0.998, 0.992, 0.998, 0.848, 0.998, 0.528],
    'F1-score': [0.998, 0.998, 0.992, 0.998, 0.848, 0.998, 0.528]
}
# 创建DataFrame
df = pd.DataFrame(data)
```

```python
import numpy as np# 设置图表参数

# 设置图表参数
plt.figure(figsize=(6, 3))

# 绘制折线图
plt.plot(df['Model'], df['Accuracy'], marker='+', label='Accuracy', color='blue', linewidth=2, alpha=0.7)
plt.plot(df['Model'], df['Precision'], marker='_', label='Precision', color='orange', linewidth=2, alpha=0.3)
plt.plot(df['Model'], df['Recall'], marker='o', label='Recall', color='green', linewidth=2, alpha=0.3)
plt.plot(df['Model'], df['F1-score'], marker='*', label='F1-score', color='red', linewidth=2, alpha=0.3)

# 设置图表属性
plt.xlabel('Model')
plt.ylabel('Scores')
plt.title('Model Performance Comparison')
plt.ylim(0, 1)  # 设置y轴范围
plt.grid()
plt.legend()
plt.tight_layout()

# 显示图表
plt.show()
```
